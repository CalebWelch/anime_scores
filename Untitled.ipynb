```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import json as js
import pandas as pd
%matplotlib inline
```

```python
anime = pd.read_json(path_or_buf='./anime_ratings')
```

```python
anime
```

```python

show =[]
genre =[]
score =[]
count =0
for i in anime:
    count+= 1
    print anime[i]
    show.append(i)
    genre.append(anime[i][0])
    score.append(anime[i][1])
```

```python
data = {'show':show,
        'genre':genre,
        'score':score}
```

```python
df = pd.DataFrame(data,index=[x for x in range(count)])
scores = [int(float(x)) for x in df.score]
```

```python
result = set(x for l in df.genre for x in l)
combo = pd.DataFrame(data=[], index=result)

for gen in result:
    scres = []
    for show, genre in df.iterrows():
        if gen in (genre[0]):
            scres.append(float(genre[1]))
        scress = np.asarray(scres)
    combo.set_value(gen,0,np.average(scress))
combo.columns = ['scores']
combo['genre'] = combo.index
print combo
```

```python
fig, ax = plt.subplots(figsize=(10,6))
fig.autofmt_xdate()
sb.countplot(x= scores,data=df,ax=ax)
sb.distplot(scores)
```
